# **Practica 1 - Regresion lineal**

```python
import statsmodels.formula.api as smf
import statsmodels.api as sm
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Descargue directamente en esta practica el csv
!wget --no-check-certificate 'https://drive.google.com/uc?export=download&id=1Khq7dnCEOSVJLtid9HVBu4jWhTzGOz_B' -O data.csv

data = pd.read_csv('data.csv')

display(data.head())
```

# **Regresion lineal simple**

```python
X1=data[['TV']]
y=data['sales']
x1=sm.add_constant(X1)
simple1=sm.OLS(y,x1).fit()
predictions1=simple1.predict(sm.add_constant(data[['TV']]))
rsel=np.sqrt(np.sum(simple1.resid**2)/(simple1.df_resid))
print(simple1.summary())
print("\n\n Error estandar residual: ",rsel)
```

Viendo los coeficientes, t estadistico, error cuadrado y ajustado y el F estadistico, vemos que este modelo simple interpreta el 61% de variavilidad de la muestra, y no esta mal su estadistico porque lo hace un modelo representativo siendo altamente significado aunque su influencia es buena no puede asegurarse que sea fiel al resultado.

```python
plt.scatter(data['TV'], data['sales'], color='blue', label='Datos reales')
plt.plot(data['TV'], predictions1, color='red', label='Regresión lineal')
plt.title('Regresión Lineal: TV vs Sales')
plt.xlabel('TV')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
# 2. Radio vs Sales
X2=data[['radio']]
y=data['sales']
x2=sm.add_constant(X2)
simple2=sm.OLS(y,x2).fit()
predictions2=simple2.predict(sm.add_constant(data[['radio']]))
rsel=np.sqrt(np.sum(simple2.resid**2)/(simple2.df_resid))
print(simple2.summary())
print("\n\n Error estandar residual: ",rsel)
```

En este este estadistico de radio haciendo las mismas consideraciones se ve que es menos significativo, no es mala si llega  ser util pero tiene aun problemas representativas ya que su estadistico es menor que la esperada, no es negativo pero solo representa un 33% aproximadamente de variavilidad

```python
plt.scatter(data['radio'], data['sales'], color='blue', label='Datos reales')
plt.plot(data['radio'], predictions2, color='red', label='Regresión lineal')
plt.title('Regresión Lineal: radio vs Sales')
plt.xlabel('radio')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
# 3. Periodico vs Sales
X3=data[['newspaper']]
y=data['sales']
x3=sm.add_constant(X3)
simple3=sm.OLS(y,x3).fit()
predictions3=simple3.predict(sm.add_constant(data[['newspaper']]))
rsel=np.sqrt(np.sum(simple3.resid**2)/(simple3.df_resid))
print(simple3.summary())
print("\n\n Error estandar residual: ",rsel)
```

El periodico es el menor en regresion simple pero se asume que aun viene siendo significativo por la t estadistica, pero es el que menos representa entre los 3 modelos incluso los ingresos son positivos pero mas adelante vemos los problemas de esta regresion simple.

```python
plt.scatter(data['newspaper'], data['sales'], color='blue', label='Datos reales')
plt.plot(data['newspaper'], predictions3, color='red', label='Regresión lineal')
plt.title('Regresión Lineal: periodico vs Sales')
plt.xlabel('radio')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

# **Regresion lineal multiple**

```python
X=data[['TV','radio','newspaper']]
y=data['sales']
x=sm.add_constant(X)
multip=sm.OLS(y,x).fit()
predictions=multip.predict(sm.add_constant(data[['TV','radio','newspaper']]))
rsel=np.sqrt(np.sum(multip.resid**2)/(multip.df_resid))
print(multip.summary())
print("\n\n Error estandar residual: ",rsel)
```

Ahora en la regresion multiple al hacer la prueba vemos que las pruebas y t estadisticos dicen lo contrario que las simples, aunque los de radio y television aun son significativos de forma positiva lo que pasa es que periodico da una representacion negativa en su estadistico entonces no contaria porque negaria la hipotesis nula de que es dependiente, entonces solo hay una forma de explicar esto, en el modelo simple se observo que tiene una correlacion con periodico y entonces dice que tiene credito en la regresion simple que tiene con el periodico entonces no se acredita del todo, por ende esta variable es la que toma credito de otra y por eso no seria una variable a considerar, si la quitamos del modelo.

```python
X=data[['TV','radio']]
y=data['sales']
x=sm.add_constant(X)
multip=sm.OLS(y,x).fit()
predictions=multip.predict(sm.add_constant(data[['TV','radio']]))
rsel=np.sqrt(np.sum(multip.resid**2)/(multip.df_resid))
print(multip.summary())
print("\n\n Error estandar residual: ",rsel)
```

Tenemos estos resultados y se adapta mucho mejor el modelo que considerando el periodico, incluso podriamos afirmar que es mas representativo y tiene mejor F estadistico el cual nos permite ver es mas demostrativo.

```python
X1=data2[['Own']]
y=data2['Balance']
x1=sm.add_constant(X1)
x1['Own'] = x1['Own'].apply(lambda x: 1 if x=='Yes' else 0)
model1=sm.OLS(y,x1).fit()
predictions1=model1.predict(sm.add_constant(data2[['Balance']]))
rsel=np.sqrt(np.sum(model1.resid**2)/(model1.df_resid))
print(model1.summary())
print("\n\n Error estandar residual: ",rsel)
```

Desde la perspectiva de nuestros estadisticos, error cuadrado y F estadistica podemos asumir que no tiene influencia las propiedades con el balance de cada usuario, en este caso se codifico con 0 si no tenia y 1 si tenia propiedad en la regresion.

```python
X1=data2[['Own']]
y=data2['Balance']
x1=sm.add_constant(X1)
x1['Own'] = x1['Own'].apply(lambda x: 1 if x=='Yes' else -1)
model1=sm.OLS(y,x1).fit()
predictions1=model1.predict(sm.add_constant(data2[['Balance']]))
rsel=np.sqrt(np.sum(model1.resid**2)/(model1.df_resid))
print(model1.summary())
print("\n\n Error estandar residual: ",rsel)
```

Cambiando la codificacion a 1 si tenia propiedad y -1 si no tenia sigue mostrando lo mismo en estadisticos lo que cambia solo es en el t estadistico y el coeficicente absorve mas del ingreso fijo, pero sigue mostrando la misma influencia no es representativa esta consideracion para el balande de un usuario

```python
data2['South'] = data2['Region'].apply(lambda x: 1 if x == 'South' else 0)
data2['West'] = data2['Region'].apply(lambda x: 1 if x == 'West' else 0)
X = data2[['South', 'West']]
X = sm.add_constant(X)
y = data2['Balance']
model = sm.OLS(y, X).fit()
predictions = model.predict(X)
rsel = np.sqrt(np.sum(model.resid**2) / model.df_resid)
print(model.summary())
print("\n\n Error estándar residual: ", rsel)
```

En este estadistico vemos que igual es un caso completamente negativo incluso las pruebas e hipotesis de ambas variables dicen que no son aceptadas entonces podemos asumir por los resultados que de plano no es un modelo que sirva para el balance incluso nuestro F estadistico es menor a 1 entonces no nos sirven estas variables para asumir un balance crediticio ya que no tiene una significancia.

```python
data['TV_x_Radio']=data['TV']*data['radio']
X=data[['TV','radio','TV_x_Radio']]
y=data['sales']
x=sm.add_constant(X)
model=sm.OLS(y,x).fit()
predictions=model.predict(sm.add_constant(data[['TV','radio','TV_x_Radio']]))
print(model.summary())
```

```python
# Crear una malla de valores para TV y radio
tv_range = np.linspace(data['TV'].min(), data['TV'].max(), 100)
radio_range = np.linspace(data['radio'].min(), data['radio'].max(), 100)
tv_mesh, radio_mesh = np.meshgrid(tv_range, radio_range)
interaction_mesh = tv_mesh * radio_mesh
X_mesh = sm.add_constant(np.column_stack((tv_mesh.ravel(), radio_mesh.ravel(), interaction_mesh.ravel())))
sales_pred = model.predict(X_mesh).reshape(tv_mesh.shape)

# Crear el gráfico de superficie
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(tv_mesh, radio_mesh, sales_pred, cmap='viridis', alpha=0.6)

# Añadir los puntos reales
ax.scatter(data['TV'], data['radio'], data['sales'], color='red', label='Datos reales')

# Etiquetas y título
ax.set_xlabel('TV')
ax.set_ylabel('Radio')
ax.set_zlabel('Sales')
ax.set_title('Superficie de Predicción de Sales en función de TV y Radio')
plt.legend()
plt.show()
```

El modelo con una dependencia cruzada de lo que seria tv x radio hace que sea mas preciso el modelo, sube su representacion al 96% y su F estadistico al ser mucho mayor hacer que sea mas complejo pero tiene mejor descriptivilidad, pero se ve que los coeficientes y t estadisticos dicen que el producto de TV_x_Radio no es representativo pero si es importante ya que al hacer una regresion lineal dentro de una variable hace que este valor cruzado sea mas fiel y no podemos ignorarlo ya que acredita que estos atributos si impactan de forma positiva la influencia a las ventas

```python
!wget --no-check-certificate 'https://drive.google.com/uc?id=1k3qcS5wam19fQi65wIfC_Tdubx1hu6yN&export=download' -O data3.csv

data3 = pd.read_csv('data3.csv')

display(data3)
```

```python
data3['horsepower'] = pd.to_numeric(data3['horsepower'], errors='coerce')
data3 = data3.dropna(subset=['horsepower'])
data3['horsepower_2']=data3['horsepower']**2
X = data3[['horsepower','horsepower_2']]
X = sm.add_constant(X)
y = data3['mpg']
model = sm.OLS(y, X).fit()
predictions = model.predict(X)
rsel = np.sqrt(np.sum(model.resid**2) / model.df_resid)
print(model.summary())
print("\n\n Error estándar residual: ", rsel)
```

```python
# Gráfica
plt.figure(figsize=(10, 6))
plt.scatter(data3['horsepower'], data3['mpg'], alpha=0.5, label='Datos observados')
plt.scatter(data3['horsepower'], predictions, color='red', alpha=0.5, label='Predicciones del modelo')
plt.xlabel('Horsepower')
plt.ylabel('MPG')
plt.title('Relación no lineal entre Horsepower y MPG')
plt.legend()
plt.show()
```

El modelo considerando estas dos variables de horsepower y su cuadrado tiene una variavilidad del 68% aproximadamente y una significancia por su F estadistico, vemos que cuando no hay influencia de las potencias esta ya tiene un valor fijo elevado pero dependiendo de la mpg bajara ya que tiene una influencia negativa con la potencia, pero al tener el cuadrado nos permite hacer una curvatura la cual hace casi nula, es una no linealidad el cual causaria que mpg se reduciera gradualmente pero con este ajuste permite capturar que el impacto de la potencia no es constante y aumenta a medida de esta misma potencia.
